```python
# function library for final project "Methane"


def max_ppb_per_month(data_array, month_array): # function for finding monthly max ppb values
# data_array is ppb
# month_array is month

  max_ppb_by_month = [] # stores max ppb for each month (1 to 12)
  for x in range(1, 13):  # for each month from 1 to 12
    month_ppb = []

    for i in range(len(data_array)): # collect all ppb values for the current month
      if month_array[i] == x:
        month_ppb.append(data_array[i])

    # finding max ppb values for the month
    if len(month_ppb) > 0:
      max_value = month_ppb[0]
      for value in month_ppb:
        if value > max_value:
          max_value = value
      max_ppb_by_month.append(max_value)
    else:
      max_ppb_by_month.append(None)  # if month has no data

  return max_ppb_by_month


def avg_ppb_by_month(ppb_array, month_array):
    monthly_avg = []

    for m in range(1, 13):
        month_ppb = []  # index all the month values

        for i in range(len(month_array)):
            if month_array[i] == m:
                month_ppb.append(ppb_array[i])

        if len(month_ppb) > 0: # calc monthly average
            total = 0
            for value in month_ppb:
                total = total + value
            average = total / len(month_ppb) # data calculation for average
        else:
            average = None  # if month has no data

        monthly_avg.append(average)

    return monthly_avg





```
